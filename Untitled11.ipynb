```
import pandas as pd
```

```
#Load Dataset. Dataset consists of 8 variables which are used to classify into whether the person will get diabates in the next 5 years or not.
df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv',sep = ',')

#Naming the columns of the dataset.
df.columns=['Number of times pregnant','Plasma glucose concentration','Diastolic blood pressure',
            'Triceps skinfold thickness','2-Hour serum insulin','Body mass index','Diabetes pedigree function',
          'Age','Class']
```

```
#Preview of dataset
df
```

```
#Balance the dataset
m=0
n=0
i=0
for i in range(0,767):
  if df.iloc[i,8]==0:
    m+=1
  else:
    n+=1
print(m)
print(n)
```

```
# 0 is majoirity class and 1 is minority class. We need to balance the dataset.
import sklearn

df_majority = df[df.Class.values==0]
df_minority = df[df.Class.values==1]
 
# Downsample majority class
df_majority_downsampled = sklearn.utils.resample(df_majority, 
                                 replace=False,    # sample without replacement
                                 n_samples=267,     # to match minority class
                                 random_state=123) # reproducible results
 
# Combine minority class with downsampled majority class
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
df_downsampled.Class.value_counts()
```
